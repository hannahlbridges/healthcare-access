```python
import pandas as pd
```

```python
median_income_df = pd.read_csv('income_and_spending_data/median_income_2018.csv')

expansion_dates_df = pd.read_csv("medicaid_data/medicaid_expansion_dates.csv")



state_spending_df = pd.read_csv('income_and_spending_data/usgs_state_2018.csv')




expansion_dates_df = expansion_dates_df.rename(columns={"Date Implemented":"Date Added"})
expansion_dates_df = expansion_dates_df.rename(columns={"State":"State"})

expansion_dates_df

median_income_df = median_income_df.rename(columns={"State ":"State"})
median_income_df


state_spending_df = state_spending_df.rename(columns={"Comparison of State and Local Government Spending":"State"})
state_spending_df = state_spending_df.drop([0,1,2,3,4,13])
state_spending_df = state_spending_df.rename(columns={"Unnamed: 1":"State Spending", "Unnamed: 2":"Local Spending", "Unnamed: 3":"State and Local Spending", "Unnamed: 4":"Gross State Product", "Unnamed: 5":"Real State Growth Percentage", "Unnamed: 6":"Population (Millions)"})
state_spending_df
```
