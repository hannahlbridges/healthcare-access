```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
avg_cost_ha_2013 = pd.read_csv("medicare_disparities/2013_average_pcost_heart_attack.csv")
avg_cost_ha_2013
```

```python
avg_cost_ha_2013.rename({'analysis_value':'avg_cost_ha_2013'}, axis=1, inplace=True)
avg_cost_ha_2013
```

```python
ha_2013 = avg_cost_ha_2013[['state','avg_cost_ha_2013']]
ha_2013
```

```python
avg_cost_stroke_2013 = pd.read_csv("medicare_disparities/2013_average_pcost_stroke.csv")
avg_cost_stroke_2013
```

```python
avg_cost_stroke_2013.rename({'analysis_value':'avg_cost_stroke_2013'}, axis=1, inplace=True)
```

```python
medicare_avg_costs2 = ha_2013.merge(stroke_2013, on='state')
```

```python
medicare_avg_costs2
```
