```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import pearsonr 
from scipy.stats import linregress
%matplotlib inline
```

```python
state_data = pd.read_csv("final_dfs/state_data.csv")
state_data
```

```python
coverage = pd.read_csv('final_dfs/coverage_type.csv')
coverage.head()
```

```python
labels = ['Marketplace/Private', 'Medicaid', 'Medicare', 'Military','Uninsured']
sizes = coverage[['Marketplace/Private', 'Medicaid', 'Medicare', 'Military', 'Uninsured']].iloc[33]
plt.pie(sizes, labels=labels, colors=['green', 'blue', 'purple', 'red'], autopct="%1.1f%%", startangle=45,)
plt.title('Percent of North Carolina Citizens by Coverage Type')
plt.legend(loc='lower right')
plt.savefig('coverage_type.png')
```

```python
x_values = state_data['State and Local Spending']
y_values = state_data['2018 Uninsured']

plt.scatter(x_values, y_values, marker='o', facecolors='green', edgecolors='blue', alpha=.75)
plt.ylabel('Percent Uninsured')
plt.xlabel('State and Local Spending')
plt.title('2018 State Spending Compared to the Percent of Uninsured')

plt.show()
plt.savefig('spending_uninsured.png')
```
