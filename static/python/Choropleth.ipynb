```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("MEPS_data/AvgSinglePremium.csv")
```

```python
df
df.drop([7,44])
```

```python
fig = go.Figure(data=go.Choropleth(
    locations = df['State'],
    locationmode = 'USA-states',
    z = df['2018'],
    colorscale ='Viridis',
    colorbar_title = "Healthcare Plan Premiums"
))
fig.update_layout(
    title_text = "Healthcare Plan Premiums",
    geo_scope = 'usa')
fig.show()
```
